```python

df.columns = df.columns.str.strip()

df = df.rename(columns={
    'Aptitude student score(30 Marks)': 'Aptitude',
    'Logical Reasoning student score(30 Marks)': 'Reasoning',
    'Verbal student score(20 Marks)': 'Verbal',
    'Technical MCQs student score(30 Marks)': 'Tech_MCQ',
    'Psuedocode student score(10 Marks)': 'Pseudocode',
    'Coding student score(30 Marks)': 'Coding',
    'Student score(150 Marks)': 'Total'
})


score_col = ['Aptitude', 'Reasoning', 'Verbal', 'Tech_MCQ', 'Pseudocode', 'Coding', 'Total']


df[score_col] = df[score_col].apply(pd.to_numeric, errors='coerce')


avg_score = df[score_col].mean()


print("Average Score in Each Subject:")
print(avg_score)
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
avg_score.plot(kind='bar', color='red', edgecolor='black')

plt.title('Average Score in Each Subject')
plt.ylabel('Average Marks')
plt.xlabel('Subjects')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df.columns = df.columns.str.strip()


df = df.rename(columns={
    'Aptitude student score(30 Marks)': 'Aptitude',
    'Logical Reasoning student score(30 Marks)': 'Reasoning',
    'Verbal student score(20 Marks)': 'Verbal',
    'Technical MCQs student score(30 Marks)': 'TechMCQ',
    'Psuedocode student score(10 Marks)': 'Pseudocode',
    'Coding student score(30 Marks)': 'Coding'
})


max_s = {'Aptitude': 30, 'Reasoning': 30, 'Verbal': 20,
              'TechMCQ': 30, 'Pseudocode': 10, 'Coding': 30}


below_50 = {}
for col, max_mark in max_s.items():
    df[col] = pd.to_numeric(df[col], errors='coerce')
    below_50[col] = (df[col] < (0.5 * max_mark)).sum()

pd.Series(below_50).plot(kind='bar', color='mediumseagreen', edgecolor='black')
plt.title('Students Scoring Below 50% in Each Section')
plt.ylabel('Number of Students')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
